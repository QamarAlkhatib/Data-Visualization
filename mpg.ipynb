## Load the data

```python
data = sns.load_dataset('mpg')
```

## a sample data 

```python
data.head(5)
```

## Ploting using seaborn example gallery

### mpg for origin usa, japan, europe according to the displacement and model year. 

```python
sns.set_theme(style="ticks")

colors = (250, 70, 50), (350, 70, 50)
cmap = sns.blend_palette(colors, input="husl", as_cmap=True)
sns.displot(
    data,
    x="displacement", col="origin", hue="model_year",
    kind="ecdf", aspect=.75, linewidth=2, palette=cmap,
)
```

### The horsepower vs the acceleration for each mpg

```python
g = sns.JointGrid(data=data, x="mpg", y="acceleration", space=0, ratio=17)
g.plot_joint(sns.scatterplot, size=data["horsepower"], sizes=(30, 120),color="g", alpha=.6, legend=False)
g.plot_marginals(sns.rugplot, height=1, color="g", alpha=.6)
```

### The horsepower for the mpg in the usa vs japan vs europe

```python
sns.relplot(x="horsepower", y="mpg", hue="origin", size="weight",
            sizes=(40, 400), alpha=.5, palette="muted",
            height=6, data=data)
```
