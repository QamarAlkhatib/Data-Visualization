# Data Visualization For planets

#### import the seaborn

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

#### Set default plotter to be seaborn

```python
sns.set()
```

#### show the data seaborn names to chose from

```python
sns.get_dataset_names()
```

## Load the data

```python
data = sns.load_dataset('planets')
```

## a sample data 

Here in the data we have different types of planets method,number,orbital period,mass,distance and year

```python
data.head(10)
```

## Ploting using seaborn example gallery 

### Plot the planets orbit method with the distance

```python
sns.set_theme(style="ticks")
f, ax = plt.subplots(figsize=(10, 9))
ax.set_xscale("log")

sns.boxplot(x="distance", y="method", data=data,
            whis=[0, 100], width=.9, palette="vlag")

sns.stripplot(x="distance", y="method", data=data,
              size=4, color=".3", linewidth=0)
```

### The planet mass size according to its distance,discovery year and orbital period

```python
sns.set_theme(style="darkgrid")
cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)
g = sns.relplot(
    data=data,
    x="distance", y="orbital_period",
    hue="year", size="mass",
    palette=cmap, sizes=(10, 200),
)
g.set(xscale="log", yscale="log")
g.ax.xaxis.grid(True, "minor", linewidth=.25)
g.ax.yaxis.grid(True, "minor", linewidth=.25)
g.despine(left=True, bottom=True)
```

### For each planet the distance according the discovery year

```python
sns.set_theme(style="ticks")

g = sns.JointGrid(data=data, x="year", y="distance", marginal_ticks=True)
g.ax_joint.set(yscale="log")
cax = g.figure.add_axes([.15, .55, .02, .2])
g.plot_joint(
    sns.histplot, discrete=(True, False),
    cmap="light:#03012d", pmax=.8, cbar=True, cbar_ax=cax
)
g.plot_marginals(sns.histplot, element="step", color="#03012d")
```

Summery:

- Most of the planets discovery year were in 2005-2010.
